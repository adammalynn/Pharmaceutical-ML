```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read the csv file into a pandas DataFrame
lsd = pd.read_csv('resources/lumab_stab_scrubbed.csv')
lsd.head()
```

```python
lsd.count()
```

```python
lsd.dtypes
```

```python
convert = lsd
convert.head()
```

```python
# delete first column
convert=convert.drop(convert.columns[0], axis=1)
```

```python
# strip off M from timepoint
convert['TIMEPOINT'] = convert['TIMEPOINT'].str.extract('(\d+)', expand=False)
convert.head()
```

```python
convert.dtypes
```

```python
# convert timepoint to integer
convert['TIMEPOINT'] = convert['TIMEPOINT'].astype(int)
```

```python
convert.dtypes
```

```python
# strip < from entry
convert['ENTRY'] = convert['ENTRY'].str.replace('<', '')

```

```python
# convert entry to float
# Using to_numeric() to convert a column's data into floats
convert["ENTRY"] = pd.to_numeric(
    convert["ENTRY"])
convert.dtypes

#convert['ENTRY'] = convert['ENTRY'].astype(float)
```

```python
convert.dtypes
```

```python
convert.count()
```

```python
# remove the rows with any missing data
convert = convert.dropna(how='any')
```

```python
convert.count()
```

```python
convert
```

```python
# round to 2decimal places
convert["ENTRY"] = convert["ENTRY"].map("{:.2f}".format)
```

```python
convert.dtypes
```

```python
# convert entry to float

# Using to_numeric() to convert a column's data into floats
convert["ENTRY"] = pd.to_numeric(
    convert["ENTRY"])
convert.dtypes

# convert['ENTRY'] = convert['ENTRY'].astype(float)
```

```python
convert.replace(r'/\w+',value="",inplace=True,regex=True)
convert['Lot'].replace(r'[A-Z,a-z]+',value="",inplace=True,regex=True)
```

```python
convert.dtypes
```

```python
# convert Lot to int

# Using to_numeric() to convert a column's data into floats
convert["Lot"] = pd.to_numeric(
    convert["Lot"])
convert.dtypes
```

```python
df1 = convert.loc[convert["CONDITION"] == "5C", :]
```

```python
df2 = convert.loc[convert["CONDITION"] == "25_60", :]
```

```python
df2.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
df1.to_csv("Output/lumid_5c.csv", index=False, header=True)
```
