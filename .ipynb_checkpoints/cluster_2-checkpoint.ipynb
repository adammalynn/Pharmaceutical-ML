```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
#import csv file
file = 'Output/clean_zumacin.csv'
```

```python
df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
df.head()
```

```python
#drop null values in the assay date column
df = df.dropna(subset=['Assay_Date'])
```

```python
df.fillna(value=0)
```

```python
df.dtypes
```

```python
df=df.drop(['Chemist', 'Lab_Site'], axis=1)
```

```python
# plt.scatter(df[:, 1], df[:, 15])
X=df.iloc[:, 1]
y=df.iloc[:, 15]
```

```python
#set the y-axis limit to hide the null(0) values
y_limit=7
y_axis=np.arange(1, y_limit, 1)
```

```python
plt.scatter(X,y, edgecolors="black")
plt.title("Aggregate % Over Time")
plt.ylabel('Agg%')
plt.xlabel('Assay Date')
plt.ylim(1, y_limit)

plt.figure(figsize=(30,20))
```

```python
plt.savefig("Output/Agg%OverTime.png")
```
