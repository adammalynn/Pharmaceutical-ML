```python
organized_df = df[["Agg%","Main%"]]
organized_df.head()
```

```python
organized_df.dropna(inplace=True)
```

```python
organized_df.head()
```

```python
data_array=np.array(organized_df)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
```

```python
# Fit the model to the data
kmeans.fit(data_array)
```

```python
# Predict the clusters
predicted_clusters = kmeans.predict(data_array)
```

```python
mpl.rcParams['figure.figsize'] = [14,8]
```

```python
# Plot the predicted clusters to see if the model predicted the correct clusters
# This is visual validation that the model was trained correctly.
plt.scatter(data_array[:, 0], data_array[:, 1], c=predicted_clusters, s=50, cmap='viridis')
plt.title("Purity %", fontsize =20)

plt.savefig("Output/Purity.png")
```
