```python
# Read the csv file into a pandas DataFrame
dfs = pd.read_csv('Output/lumid_5c.csv')
dfs.head()
```

```python
# Obtain the x and y coordinates for each of the Lot types
#Split up our data into groups based upon 'Lot' and 'Sample-Number' then perform calculations
lot_type = dfs.groupby(["Lot", "Sample-Number"]).agg({"TIMEPOINT":["mean"] , "ENTRY":["sum"]})

#name the columns of the calculated values
lot_type.columns = ["time", "measurement"]

lot_type.head()
```

```python
# Assign the data to X and y
# Note: Sklearn requires a two-dimensional array of values
# so we use reshape to create this

X = xa_axis.values.reshape(-1, 1)
y = ya_axis.values.reshape(-1, 1)
x123 = full_datax.values.reshape(-1,1)
y123 = full_datay.values.reshape(-1,1)

print("Shape: ", X.shape, y.shape)
print(x123.shape)
```

```python
lot_type.count()
```

```python
# Use sklearn's `train_test_split` to split the data into training and testing

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
X1_train, X1_test, y1_train, y1_test = train_test_split(x123, y123, random_state=1)
```

```python
# Create the model and fit the model to the data

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model1 = LinearRegression()
```

```python
# Fit the model to train the data. 
# This is the training step where you fit the line to the data.

model.fit(X_train, y_train)
model1.fit(X1_train, y1_train)
```

```python
# Print the coefficient and the intercept for the model
print('Weight coefficients: ', model.coef_)
print('y-axis intercept: ', model.intercept_)

# Print the coefficient and the intercept for the model
print('Weight coefficients: ', model1.coef_)
print('y-axis intercept: ', model1.intercept_)
```

```python
# Note: we have to transform our min and max values 
# so they are in the format: array([[ 1.17]])
# This is the required format for `model.predict()`

x_min = np.array([[X.min()]])
x_max = np.array([[X.max()]])
print(f"Min X Value: {x_min}")
print(f"Max X Value: {x_max}")
```

```python
# Calculate the y_min and y_max using model.predict and x_min and x_max

y_min = model.predict(x_min)
y_max = model.predict(x_max)
```
